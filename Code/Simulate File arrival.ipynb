```python
import time
def file_arrival(source_dir,landing_dir):
    try:
        # Get the list of all files present in the source directory 
        files = [f.path for f in dbutils.fs.ls(source_dir)]

        # Setting the batch size and interval

        batch_size = 10
        interval = 60

        index = 0

        while index < len(files):
            batch = files[index:index + batch_size]

            for file_path in batch:
                file_name = file_path.split("/")[-1]

                dbutils.fs.cp(file_path,f"{landing_dir}/{file_name}")

                print(f"Copied: {file_name}")

            index += batch_size

            print("Waiting 60 seconds...\n")
            time.sleep(interval)
    except Exception as e:
        raise Exception(f"Error: {e}")


if __name__=="__main__":
    source_dir = "/Volumes/cosmetic_history/json/json_files"
    landing_dir = "/Volumes/cosmetic_history/landing/landing_zone/source_files"
    file_arrival(source_dir,landing_dir)
    print("Done!")
```
